```
import requests as r
import pandas as pd
from datetime import datetime
from google.colab import files
```

```
base_url = "https://www.wildberries.ru/brands/bosch"
# base_url = "https://www.wildberries.ru/brands/scarlett"
headers = ({
        "x-requested-with": "XMLHttpRequest"
        })
```

```
cur_datetime_str = datetime.today().strftime('%d.%m.%Y__%H-%M')
request_results = []
stop_mark = False
total_pages = 10
i = 1
while stop_mark != True and i <= total_pages:
    print(f"Parsing page #{i}")
    req = r.post(base_url, data={"page":i},headers=headers)
    if req.status_code != 200:
        stop_mark = True
        print(f"Got HHTP {req.status_code} on page #{i}")
        break

    total_pages = req.json()["pagerData"]["totalPages"]
    cur_page_df = pd.DataFrame.from_dict(req.json()["products"]).set_index("cod1S")
    request_results.append(cur_page_df)
    i += 1
    if i > total_pages:
        print(f"Work is done.")

results = pd.concat(request_results, axis=0,sort=False)
```

```
filename = 'wb_' + cur_datetime_str + '.csv'
with open(filename, 'w') as f:
    # https://stackoverflow.com/a/18239236
    f.write('\ufeff' + results.to_csv(sep=";", decimal=","))

files.download(filename)
```
